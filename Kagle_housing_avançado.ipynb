```python
import pandas as pd

pd.set_option('display.max_columns',81)
pd.set_option('display.max_rows',81)
```

```python
treino = pd.read_csv(r'./kaggle_files/train.csv')
print('Dim of dataset',
      '\nColunas/Variaveis: ',treino.shape[1], '\n'
      'Linhas: ', treino.shape[0])
print('-'*60)
display(treino.head(3))
display(treino.tail(3))
```

```python
(treino.isnull().sum()/treino.shape[0]).sort_values(ascending=False)[:21]
```

```python
excluir = treino.columns[treino.isnull().sum()/treino.shape[0] >0.1]
excluir
```

```python
treino = treino.drop(excluir,axis=1)
```

```python
treino = treino.fillna(-1)
```

```python
(treino.isnull().sum()/treino.shape[0]).sort_values(ascending=False)[:3]
```

```python
col_nr = treino.columns[treino.dtypes != 'object']
```

```python
#removendo colunas textos para primeira analise
treino = treino.loc[:,col_nr]
```

```python
X = treino.drop('SalePrice',axis=1)
y= treino.SalePrice
```

### Separando o modelo

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

### Escolhendo algoritimos mais simples para analises iniciais
- Regressão linear
- Arvore de Regressão
- KNeighborsRegressor

```python
#importando a Regressao linear
from sklearn.linear_model import LinearRegression
```

```python
# criando e fazendo o fi com os dados de treino
reg_rl = LinearRegression().fit(X_train,y_train)
```

```python
# Fazendo a previsão pros dados de teste
y_pred_rl = reg_rl.predict(X_test)
```

```python
# Importando a árvore de regressão
from sklearn import tree

# Criando o regressor e fazendo o fit com os dados de treino
reg_ad = tree.DecisionTreeRegressor().fit(X_train, y_train)

# Fazendo a previsão
y_pred_ad = reg_ad.predict(X_test)
```

```python
# Importando o KNN
from sklearn.neighbors import KNeighborsRegressor

# Criando o regressor e fazendo o fit com os dados de treino
reg_knn = KNeighborsRegressor(n_neighbors=2).fit(X_train, y_train)

# Fazendo a previsão
y_pred_knn = reg_knn.predict(X_test)
```

```python
# Importando o erro médio absoluto
from sklearn.metrics import mean_absolute_error
# E o erro quadrático médio
from sklearn.metrics import mean_squared_error
```

```python
# Avaliando o erro da regressão
print(mean_absolute_error(y_test, y_pred_rl))
print(mean_squared_error(y_test, y_pred_rl))
```

```python
# da árvore de decisão
print(mean_absolute_error(y_test, y_pred_ad))
print(mean_squared_error(y_test, y_pred_ad))
```

```python
# e do knn
print(mean_absolute_error(y_test, y_pred_knn))
print(mean_squared_error(y_test, y_pred_knn))
```
